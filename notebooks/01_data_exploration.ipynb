# Data Exploration and Preprocessing
In this notebook, we explore the MNIST dataset, view the data distribution, and visualize some sample images.
We also apply our preprocessing pipeline from `src.data_loader` and our augmentation pipeline from `src.data_augmentation`.

```
import sys
import os
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Add project root to path so we can import src
sys.path.append(os.path.abspath('..'))

from src.data_loader import load_mnist_data, normalize_images, reshape_for_cnn, create_train_val_split
from src.data_augmentation import visualize_augmentation
```

## Load Data

```
(X_train, y_train), (X_test, y_test) = load_mnist_data()
print(f"X_train shape: {X_train.shape}, y_train shape: {y_train.shape}")
print(f"X_test shape: {X_test.shape}, y_test shape: {y_test.shape}")
```

## Data Distribution

```
plt.figure(figsize=(10, 5))
sns.countplot(x=y_train)
plt.title("Distribution of Digits in Training Data")
plt.xlabel("Digit")
plt.ylabel("Count")
plt.show()
```

## Visualize Sample Images

```
fig, axes = plt.subplots(2, 5, figsize=(12, 5))
axes = axes.flatten()

for i in range(10):
    # Find first index of digit i
    idx = np.where(y_train == i)[0][0]
    axes[i].imshow(X_train[idx], cmap='gray')
    axes[i].set_title(f"Label: {i}")
    axes[i].axis('off')
    
plt.tight_layout()
plt.show()
```

## Preprocessing Pipeline

```
X_train_norm = normalize_images(X_train)
X_test_norm = normalize_images(X_test)

X_train_cnn = reshape_for_cnn(X_train_norm)
X_test_cnn = reshape_for_cnn(X_test_norm)

X_t, X_v, y_t, y_v = create_train_val_split(X_train_cnn, y_train, val_split=0.2)
print("After splitting:")
print(f"Training data: {X_t.shape}")
print(f"Validation data: {X_v.shape}")
print(f"Testing data: {X_test_cnn.shape}")
```

## Augmentation Pipeline

```
sample_img = X_train[0]
aug_dict = visualize_augmentation(sample_img)

fig, axes = plt.subplots(1, len(aug_dict), figsize=(18, 4))
for ax, (name, img) in zip(axes, aug_dict.items()):
    ax.imshow(img, cmap='gray')
    ax.set_title(name)
    ax.axis('off')
    
plt.tight_layout()
plt.show()
```
