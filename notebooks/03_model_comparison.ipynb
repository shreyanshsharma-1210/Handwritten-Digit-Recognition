# Model Comparison Analysis
In this notebook, we analyze the performance of our three trained models:
1. **Simple Neural Network**
2. **Basic CNN (LeNet-style)**
3. **Advanced CNN (BatchNorm + Dropout)**

We compare them based on accuracy, parameter count, model size, and inference speed.

```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from IPython.display import display, Image
```

## Comparison Results

```
# Load the comparison CSV
df = pd.read_csv('../results/reports/model_comparison.csv')
display(df)
```

## Visualizing Performance

```
# Visualize Accuracy
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Accuracy', data=df, palette='viridis')
plt.title('Model Accuracy Comparison')
plt.ylim(0.95, 1.0)
plt.show()
```

```
# Visualize Inference Speed
plt.figure(figsize=(10, 6))
sns.barplot(x='Model', y='Inference Time (ms/sample)', data=df, palette='magma')
plt.title('Inference Latency Comparison')
plt.show()
```

### Key Findings:
1. **Accuracy**: The **Basic CNN** actually achieved the highest accuracy on the test set (~99.2%).
2. **Efficiency**: The **Advanced CNN** is the most compact (fewest parameters and smallest file size) but has a higher inference latency due to its deeper structure and batch normalization layers.
3. **Trade-offs**: The **Simple NN** is extremely fast but slightly less accurate than the CNN models.
